```python
unknown_summary = {}

for col in df_raw.select_dtypes(include="object").columns:
    n_unknown = (df_raw[col] == 'unknown').sum()
    if n_unknown > 0:
        unknown_summary[col] = n_unknown
        print(f"{col}: {n_unknown} unknown ({n_unknown/len(df_raw):.1%})")
```

```python
cols_to_check = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for col in cols_to_check:
    print(f"\n{'='*45}")
    print(f"  Column: {col.upper()}")
    print(f"{'='*45}")
    
    counts    = df_raw[col].value_counts()
    percents  = df_raw[col].value_counts(normalize=True) * 100
    
    summary = pd.DataFrame({
        'Category'   : counts.index,
        'Count'      : counts.values,
        'Percentage' : percents.values.round(2)
    })
    summary['Percentage'] = summary['Percentage'].apply(lambda x: f"{x}%")
    
    print(summary.to_string(index=False))
```

Treatments for Data Quality Issues

```python
#dropping duplicates
df_raw= df_raw.drop_duplicates().reset_index(drop=True)

df_raw.shape
```

```python
# Mode imputation for job, marital, housing, loan columns as unknown percentage is comparatively less
for col in ['job', 'marital', 'housing', 'loan']:
    mode_val = df_raw[col][df_raw[col] != 'unknown'].mode()[0]
    unknown_count = (df_raw[col] == 'unknown').sum()
    
    df_raw[col] = df_raw[col].replace('unknown', mode_val)
    
    print(f"Column : {col}")
    print(f"Mode   : '{mode_val}'")
    print(f"Rows imputed : {unknown_count}")
    print(f"Remaining 'unknown' : {(df_raw[col] == 'unknown').sum()}")
    print("-" * 35)
```

```python
# Group-by mode imputation for Education based on Job
mode_education_per_job = df_raw[df_raw['education'] != 'unknown'].groupby('job')['education'].agg(lambda x: x.mode()[0])

def impute_education(row):
    if row['education'] == 'unknown' and row['job'] in mode_education_per_job:
        return mode_education_per_job[row['job']]
    return row['education']

df_raw['education'] = df_raw.apply(impute_education, axis=1)
```
